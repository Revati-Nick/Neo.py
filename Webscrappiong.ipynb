```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

my_url = 'https://covidlive.com.au/report/daily-hospitalised/nsw'

uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup =soup(page_html, "html.parser")

containers = page_soup.findAll("tr", {"class": "even"})
container=containers[0]

Date_container = container.findAll("tr",{"class": "COL1 DATE"})
HOSP_container = container.findAll("td",{"class": "COL2 HOSP"})
ICU_container = container.findAll("td",{"class": "COL3 HOSP"})
VENT_container = container.findAll("td",{"class": "COL4 HOSP"})

filename="product.csv"
f= open(filename,"w")

headers="Date,HOSP,ICU,VENT\n"
f.write(headers)

for container in containers:
    Date_container = container.findAll("td",{"class": "COL1 DATE"})
    Date = Date_container[0].text.strip()
    
    HOSP_container = container.findAll("td",{"class": "COL2 HOSP"})
    HOSP = HOSP_container[0].text 
    
    ICU_container = container.findAll("td",{"class": "COL3 ICU"})
    ICU = ICU_container[0].text.strip()
    
    VENT_container = container.findAll("td",{"class": "COL4 VENT"})
    VENT = VENT_container[0].text.strip()
    
    
    print("Date:" + Date)
    print("HOSP:" + HOSP)
    print("ICU:" + ICU)
    print("VENT:" + VENT)

f.close()
```
